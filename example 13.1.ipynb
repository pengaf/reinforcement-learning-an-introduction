```python
import numpy as np
import matplotlib.pyplot as plt
import math

def short_corridor_true_value(right_prob):
    rp = right_prob
    lp = 1.0 - rp
    order = 3
    i = np.identity(order)
    t = np.array([
        [lp, rp, 0],
        [rp, 0, lp],
        [0, lp, 0],
    ])
    b = np.array([-1,-1,-1])
    m = i - t
    s = np.linalg.solve(m,b)
    return s

vs = []
for n in range(100):
    prob = (n+2)/102
    vs.append(short_corridor_true_value(prob)[0])

plt.plot(range(len(vs)), vs)
```
